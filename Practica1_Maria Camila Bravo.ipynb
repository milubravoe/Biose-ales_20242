```python
# Importar bibliotecas:
import numpy as np
import matplotlib.pyplot as plt
```

```python
# a) Crear el par de vectores
a = np.array([3.1, 1, -0.5, -3.2, 6])
b = np.array([1, 3, 2.2, 5.1, 1])
```

```python
# b) Implementar la multiplicación escalar (producto punto) de a.b
# La multiplicación escalar se puede realizar usando np.dot()
dot_product = np.dot(a, b)
print(f"Producto punto de a y b: {dot_product}")
```

```python
# c) Implementar la multiplicación punto a punto (element-wise) de a.b
# La multiplicación punto a punto se realiza usando el operador *
elementwise_product = a * b
print(f"Multiplicación punto a punto de a y b: {elementwise_product}")
```

```python
# d) Construir la matriz A
A = np.array([
    [2, -1, -3],
    [4, 1.5, -2.5],
    [7.3, -0.9, 0.2]
])
print(f"Matriz A:\n{A}")
```

```python
# e) Obtener la transpuesta de la matriz A
A_T = A.T
print(f"Transpuesta de la matriz A:\n{A_T}")
```

```python
# f) Consultar la función de los siguientes comandos de la librería numpy  y ejemplificar su uso

# ones: Crea un array lleno de unos:
ones_array = np.ones((2, 3))
print(f"Array de unos:\n{ones_array}")

# round: Redondea los valores a la cantidad de decimales especificada
rounded_array = np.round(A, 2)
print(f"Array redondeado:\n{rounded_array}")

# ceil: Redondea los valores hacia arriba
ceil_array = np.ceil(A)
print(f"Array con valores redondeados hacia arriba:\n{ceil_array}")

# floor: Redondea los valores hacia abajo
floor_array = np.floor(A)
print(f"Array con valores redondeados hacia abajo:\n{floor_array}")
```

```python
# g) Acceder al valor de la primera fila, tercera columna de la matriz A
value = A[0, 2]
print(f"Valor en la primera fila, tercera columna de la matriz A: {value}")
```

```python
# h) Obtener la segunda fila de la matriz A
second_row = A[1, :]
print(f"Segunda fila de la matriz A: {second_row}")
```

```python
# i) Consultar el comando para conocer las dimensiones de una matriz
# Utilizar el comando .shape para obtener las dimensiones
dimensions = A.shape
print(f"Dimensiones de la matriz A: {dimensions}")
```

```python
# j) Construir la función y[n] = sin(π * 0.12n) en el intervalo 0 <= n <= 100
n = np.arange(0, 101)
y = np.sin(np.pi * 0.12 * n)
print(f"Señal y[n]:\n{y}")
```

```python
# k) Construir otra senoidal y2[n] = cos(2π * 0.03n)
y2 = np.cos(2 * np.pi * 0.03 * n)
print(f"Señal y2[n]:\n{y2}")
```

```python
# l) Generar una tercera señal que sea la suma de las dos señales y[n] y y2[n]
s = y + y2
print(f"Señal s[n] (suma de y[n] y y2[n]):\n{s}")

# cuarta señal  producto de y[n] y y2[n]
t = y * y2
print(f"Señal t[n] (producto de y[n] y y2[n]):\n{t}")
```

```python
# m) Graficar en la misma figura las señales y[n] y y2[n]
plt.figure(figsize=(12, 6))
plt.plot(n, y, label='y[n] = sin(π * 0.12n)', color='c')
plt.plot(n, y2, label='y2[n] = cos(2π * 0.03n)', color='m')
plt.xlabel('n')
plt.ylabel('Amplitud')
plt.title('Señales y[n] y y2[n]')
plt.legend()
plt.show()
```

```python
# n) Graficar en la misma figura las señales s[n] y t[n]
plt.figure(figsize=(12, 6))
plt.plot(n, s, label='s[n] = y[n] + y2[n]', color='r')
plt.plot(n, t, label='t[n] = y[n] * y2[n]', color='m')
plt.xlabel('n')
plt.ylabel('Amplitud')
plt.title('Señales s[n] y t[n]')
plt.legend()
plt.show()
```

```python
#Reto: Escribir una función que reciba un diccionario con las notas de los alumnos de un curso y devuelva una
#serie con la nota mínima, la máxima, media y la desviación típica.

import pandas as pd
def analizar_notas(notas): #Función

    notas_array = np.array(list(notas.values())) #Ordena las notas en una lista o array.

    # Calculos sobre las notas
    min_nota = np.min(notas_array)
    max_nota = np.max(notas_array)
    media_nota = np.mean(notas_array)
    desviacion_nota = np.std(notas_array)

    # Crear una serie con pandas y los resultados
    resultados = pd.Series({
        'Nota Mínima': min_nota,
        'Nota Máxima': max_nota,
        'Nota Media': media_nota,
        'Desviación Típica': desviacion_nota
    })

    return resultados

# Ejemplo de uso
notas_dict = {
    'Maria': 85,
    'David': 90,
    'Luis': 78,
    'Rocio': 92,
    'Gabriela': 88

}

resultado = analizar_notas(notas_dict)
print("Análisis de notas:")
print(resultado)
```
